```python
from src.signal_generator import SignalGenerator
from src.data_type import Measurement
from src.data_compressor.other import CompressNTHS
from src.data_compressor.other import CompressMinMax
from src.data_compressor.other import CompressPWP
from src.data_compressor.pip import CompressPIP_ED
from src.data_compressor.pip import CompressPIP_PD
from src.data_compressor.pip import CompressPIP_VD
from src.data_compressor.paa import CompressPAA
from src.data_compressor.paa import CompressPAAVI
from src.data_compressor.paa import CompressByChunk
from src.data_compressor.pla import CompressAPCADFT
from src.data_compressor.pla import CompressSTC
from src.data_compressor.pla import CompressHigherDeriveration
```

```python
signal_generator = SignalGenerator(0, 100).sin(0.2, 0.2)
measurements_1 = [Measurement(measurement, index * 100) for index, measurement in enumerate(signal_generator.data)]

signal_generator = SignalGenerator(0, 100).with_peaks(4).sin(0.2, 0.2)
measurements_2 = [Measurement(measurement, index * 100) for index, measurement in enumerate(signal_generator.data)]

signal_generator = SignalGenerator(0, 100).with_peaks(3, direction=-1).sin(0.2, 0.2)
measurements_3 = [Measurement(measurement, index * 100) for index, measurement in enumerate(signal_generator.data)]

signal_generator = SignalGenerator(0, 100).with_peaks(4).with_peaks(3, direction=-1).sin(0.2, 0.2)
measurements_4 = [Measurement(measurement, index * 100) for index, measurement in enumerate(signal_generator.data)]
```

```python
def deep_copy(measurements):
  return [Measurement(measurement.value, measurement.timestamp) for measurement in measurements]

def make_plots(data_compressor, config):
  for index, measurements in enumerate([measurements_1, measurements_2, measurements_3, measurements_4]):
    data_compressor.set_data(deep_copy(measurements))
    data_compressor.compress()
    print(data_compressor.get_stats())
    config['fileName'] = config['fileName'] + '_' + str(index)
    data_compressor.vizualize(True, True, config)
```

```python
make_plots(CompressNTHS(), { 'title': 'NTH-item compression', 'dir': 'section2_cmp/nths', 'fileName': 'nths' })
```
